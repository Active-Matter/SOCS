```python
import numpy as np 
from matplotlib import pyplot as plt

N = 2000
dt = 0.1
k = 0.25
gamma = 0.02

x = np.zeros(N)
V = np.ones(N)

for t in range(N-1):
    x[t+1] = x[t] + V[t]*dt                        # Damped harmonic oscillator position eqution, Euler Algorithm
    V[t+1] = V[t] -gamma*V[t]*dt- k*x[t]*dt        # Damped harmonic oscillator velocity equation, Euler Algorithm

plt.figure(figsize=(15,5))
plt.plot(x)


x = np.zeros(N)
V = np.ones(N)

for t in range(N-1): 
    x[t+1] = x[t] + V[t]*dt                 # Damped harmonic oscillator position eqution, Leap-frog Algorithm
    V[t+1] = V[t] -gamma*V[t]*dt- k*(x[t]+x[t+1])/2*dt    # Damped harmonic oscillator velocity equation, Leap-frog Algorithm


plt.plot(x)
plt.legend(['Euler','Leap-frog'])
plt.ylabel('$x$')
plt.xlabel('$t$')
plt.title('Damped Harmonic oscillator')
plt.rcParams.update({'font.size': 18})
```
